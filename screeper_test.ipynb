```python
from urllib.request import urlopen
import re
from bs4 import BeautifulSoup
```

```python
url2 = "https://www.nih.gov/sitemap.xml?page=1"
page = urlopen(url2)

html = page.read().decode("utf-8")
print(html)
```

```python

start_index = 0

start_tag = "<loc>"
end_tag = "</loc>"
all_locs = []
```

```python
while True:
    start_index = html.find(start_tag, start_index)
    if start_index == -1:
        break

    start_index += len(start_tag)
    end_index = html.find(end_tag, start_index)
    if end_index == -1:
        break

    loc = html[start_index:end_index]
    all_locs.append(loc)

    start_index = end_index + len(end_tag)

print(all_locs)
print(len(all_locs))
```

```python
print(all_locs.count)
```

```python
pattern = r"<loc.*?>.*?</loc.*?>"
matches = re.findall(pattern, html, re.IGNORECASE)

all_locs = [re.sub("<.*?>", "", match) for match in matches]

print(all_locs)
print(len(all_locs))
```
